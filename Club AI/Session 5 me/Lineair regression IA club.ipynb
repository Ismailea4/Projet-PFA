```python
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import matplotlib.pylab as plt

import seaborn as sns
```

```python
Housing=pd.read_csv("Housing_price.csv")
df2=pd.read_csv("insurance_data.csv")
df3=pd.read_csv("data.csv")
```

```python
Housing.head()
```

```python
cor_matrix=Housing.corr()
plt.figure(figsize=(10,7))
sns.heatmap(cor_matrix,annot=True,cmap=plt.cm.CMRmap_r)
```

```python
LR=LinearRegression()
```

```python
X=Housing.drop(['price','semi-furnished'],axis=1)
Y=Housing['price']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
LR.fit(X_train,Y_train)
```

```python
Y_pred=LR.predict(X_test)
```

```python
Y_pred
```

```python
#Les coefficients w et b
LR.coef_
```

```python
LR.score(X_test,Y_test)
```
