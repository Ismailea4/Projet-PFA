```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("Club AI/titanic.csv")
```

```python
df
```

```python
df1=pd.get_dummies(df,columns=["Sex"],dtype=int)
df1
```

```python
X=df1.drop(["Name","index","PassengerId","Survived"],axis=1)
X
```

```python
Y=df1["Survived"]
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
models=[RandomForestClassifier(),KNeighborsClassifier(),SVC(),DecisionTreeClassifier()]
```

```python
Scores=[]
```

```python
#RandomForestClassifier

models[0].fit(X_train,Y_train)
Y_pred=models[0].predict(X_test)
score=accuracy_score(Y_pred,Y_test)
Scores.append(score)
```

```python
#KNeighborsClassifier

models[1].fit(X_train,Y_train)
Y_pred=models[1].predict(X_test)
score=accuracy_score(Y_pred,Y_test)
Scores.append(score)
```

```python
#SVC

models[2].fit(X_train,Y_train)
Y_pred=models[2].predict(X_test)
score=accuracy_score(Y_pred,Y_test)
Scores.append(score)
```

```python
#DecisionTreeClassifier

models[3].fit(X_train,Y_train)
Y_pred=models[3].predict(X_test)
score=accuracy_score(Y_pred,Y_test)
Scores.append(score)
```

```python
Scores
```

```python
print("The best model is DecisionTreeClassifier with score=",Scores[3])
print("The worst model is KNeighborsClassifier with score=",Scores[1])
```
