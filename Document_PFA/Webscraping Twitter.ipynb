```python
import pandas as pd
from tqdm.notebook import tqdm
```

```python
from ntscraper import Nitter
```

```python
scraper2 = Nitter()
```

```python
tweets = {'threads': [], 'tweets': []}
while len(tweets['tweets'])<500:
    tweets = scraper2.get_tweets("ww3uptodate",mode='user',number=1000)
```

```python
from pprint import pprint
```

```python
pprint(tweets)
```

```python
tweets['tweets'][0].keys()
```

```python
Tweetdata = {
    'profile_id' : [],
    'username' : [],
    'date' : [],
    'link' : [],
    'text' : [],
    'comments' : [],
    'likes' : [],
    'quotes' : [],
    'retweets' : [],
    
}       
```

```python
for tweet in tweets['tweets']:
    if tweet['text']!='':
        Tweetdata['profile_id'].append(int(tweet['user']['profile_id']))
        Tweetdata['username'].append(tweet['user']['username'])
        Tweetdata['date'].append(tweet['date'])
        Tweetdata['link'].append(tweet['link'])
        Tweetdata['text'].append(tweet['text'])
        Tweetdata['comments'].append(tweet['stats']['comments'])
        Tweetdata['likes'].append(tweet['stats']['likes'])
        Tweetdata['quotes'].append(tweet['stats']['quotes'])
        Tweetdata['retweets'].append(tweet['stats']['retweets'])
```

```python
Tweetdata
```

```python
df1 = pd.DataFrame(Tweetdata)
```

```python
df2 = pd.read_csv('Palestine-Israel_tweet4.csv', index_col=0)
```

```python
df2=df2.drop('Unnamed: 0', axis=1)
```

```python
df = pd.concat([df1, df2],ignore_index=True)
```

```python
df = df.drop_duplicates()
```

```python
df['profile_id']=pd.to_numeric(df['profile_id'], downcast='integer')
```

```python
df.to_csv("Palestine-Israel_tweet5.csv")
```
